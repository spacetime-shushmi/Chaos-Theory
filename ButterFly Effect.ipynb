# Simulating Butterfly Effect

Let $\sigma$ = a ,
 $\rho$ = b ,
 $\beta$ = c
 
 
$\frac{dx}{dt} = a*(y - x)$

$\frac{dy}{dt} = x*(b - z) - y$

$\frac{dz}{dt} = xy - cz$
Equations taken from https://en.wikipedia.org/wiki/Lorenz_system

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from scipy import stats     

a = 10
b = 28
c = 8/3
dt = 0.01

def lorenz_eqt(x, y, z):
    dx = (a*(y - x))*dt
    dy = (x*(b - z) - y)*dt
    dz = (x*y - c*z)*dt
    return dx, dy, dz

interact(lorenz_eqt, x=(0, 10, 0.5), y=(0, 10, 0.5), z=(0, 10, 0.5))

# setting iteration for the coordinates, x, y and z
maximum = 5000

x = np.empty(maximum + 1)
y = np.empty(maximum + 1)
z = np.empty(maximum + 1)

# setting initial values for x, y and z
x[0] = 0
y[0] = 1
z[0] = 2


# putting x, y and z values in the function
for coordinates in range(maximum):
    dx, dy, dz = lorenz_eqt(x[coordinates], y[coordinates], z[coordinates])
# so basically, the new coordinate will be the coordinate plus the change in the coordinate to form a continuous yet not closed loop
    x[coordinates + 1] = x[coordinates] + dx
    y[coordinates + 1] = y[coordinates] + dy
    z[coordinates + 1] = z[coordinates] + dz


# plotting the 3d grap using matplotlib GCA
axes = plt.gca(projection='3d')
axes.plot(x, y, z, lw=0.7)
axes.set_xlabel("X-axis")
axes.set_ylabel("Y-axis")
axes.set_zlabel("Z-axis")
axes.set_title("ButterFly Effect- Chaos Theory")
plt.show()

```
